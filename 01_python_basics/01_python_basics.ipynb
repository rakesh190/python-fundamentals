Implement the **mean squared error** formula manually:

`MSE = (1/n) * sum((y_true - y_pred)^2)`

Given:
```python

y_pred = [2.8, 5.2, 2.0, 6.5]
```
Compute MSE without using any library (just arithmetic operators and a loop or sum).

````python
y_true = [3.0, 5.0, 2.5, 7.0]
y_pred = [2.8, 5.2, 2.0, 6.5]

# YOUR CODE HERE
````

## 3. Control Flow

````python
# if/elif/else — model selection logic
accuracy = 0.87

if accuracy >= 0.95:
    grade = "Excellent"
elif accuracy >= 0.85:
    grade = "Good"
elif accuracy >= 0.70:
    grade = "Acceptable"
else:
    grade = "Needs Improvement"

print(f"Accuracy {accuracy:.0%} → {grade}")
````

````python
# for loop — training epochs
losses = []
loss = 2.0

for epoch in range(1, 11):
    loss = loss * 0.85  # simulate decaying loss
    losses.append(round(loss, 4))
    if epoch % 2 == 0:
        print(f"Epoch {epoch:2d} | Loss: {loss:.4f}")

print(f"\nFinal loss: {losses[-1]}")
````

````python
# while loop with early stopping
best_val_loss = float('inf')
patience = 3
no_improve = 0
epoch = 0
val_losses = [0.9, 0.8, 0.75, 0.76, 0.77, 0.78]

while epoch < len(val_losses):
    val_loss = val_losses[epoch]
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        no_improve = 0
        print(f"Epoch {epoch+1}: New best val_loss = {val_loss:.3f}")
    else:
        no_improve += 1
        print(f"Epoch {epoch+1}: No improvement ({no_improve}/{patience})")
        if no_improve >= patience:
            print("Early stopping triggered!")
            break
    epoch += 1
````

## 4. Functions

````python
# Basic function with default args — common in ML APIs
def train_model(X, y, learning_rate=0.01, n_epochs=100, verbose=False):
    """Simulate model training loop."""
    loss = 1.0
    history = []
    for epoch in range(n_epochs):
        loss = loss * (1 - learning_rate)  # simplified update
        history.append(loss)
        if verbose and epoch % 20 == 0:
            print(f"  Epoch {epoch}: loss={loss:.4f}")
    return {"final_loss": round(loss, 6), "history": history}

result = train_model(X=None, y=None, learning_rate=0.05, n_epochs=50, verbose=True)
print(f"\nFinal loss: {result['final_loss']}")
````

````python
# *args and **kwargs — used in sklearn-style APIs
def evaluate(*metrics, **params):
    """Evaluate model with arbitrary metrics and params."""
    print(f"Metrics requested: {metrics}")
    print(f"Parameters: {params}")
    scores = {m: round(__import__('random').random(), 3) for m in metrics}
    return scores

scores = evaluate('accuracy', 'f1', 'roc_auc', threshold=0.5, average='macro')
print(f"Scores: {scores}")
````

````python
# Lambda functions — used in pandas apply, sorting, etc.
normalize = lambda x, min_v, max_v: (x - min_v) / (max_v - min_v)
sigmoid = lambda z: 1 / (1 + 2.718**(-z))
relu = lambda x: max(0, x)

print(f"normalize(75, 0, 100) = {normalize(75, 0, 100)}")
print(f"sigmoid(2.0) = {sigmoid(2.0):.4f}")
print(f"relu(-3) = {relu(-3)}, relu(5) = {relu(5)}")
````

### Exercise 1.4 — Functions
Write a function `compute_metrics(y_true, y_pred)` that returns a dict with:
- `accuracy`: fraction of correct predictions
- `error_rate`: 1 - accuracy
- `n_correct`: count of correct predictions
- `n_total`: total samples

Test with: `y_true = [1,0,1,1,0,1,0,0]`, `y_pred = [1,0,1,0,0,1,1,0]`

````python
# YOUR CODE HERE
````

````python
# SOLUTION
def compute_metrics(y_true, y_pred):
    """Compute basic classification metrics."""
    n_total = len(y_true)
    n_correct = sum(1 for yt, yp in zip(y_true, y_pred) if yt == yp)
    accuracy = n_correct / n_total
    return {
        "accuracy": round(accuracy, 4),
        "error_rate": round(1 - accuracy, 4),
        "n_correct": n_correct,
        "n_total": n_total
    }

y_true = [1, 0, 1, 1, 0, 1, 0, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1, 0]
metrics = compute_metrics(y_true, y_pred)
for k, v in metrics.items():
    print(f"  {k}: {v}")
````

## 6. Error Handling

Robust ML pipelines handle bad data gracefully.

````python
# try/except — handle bad input data
def safe_convert(value, dtype=float, default=None):
    """Safely convert a value, returning default on failure."""
    try:
        return dtype(value)
    except (ValueError, TypeError) as e:
        print(f"  Warning: Cannot convert '{value}' to {dtype.__name__}: {e}")
        return default

raw_data = ["1.5", "N/A", "3.2", None, "?", "4.8"]
cleaned = [safe_convert(v) for v in raw_data]
print(f"Raw:     {raw_data}")
print(f"Cleaned: {cleaned}")
````

### Exercise 1.6 — Error Handling
Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` on `ZeroDivisionError`
- Returns `None` on `TypeError` (with a warning message)

Test with: `(10, 2)`, `(5, 0)`, `("x", 2)`

````python
# YOUR CODE HERE
````

````python
# SOLUTION
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None
    except TypeError as e:
        print(f"Warning: {e}")
        return None

print(safe_divide(10, 2))
print(safe_divide(5, 0))
print(safe_divide("x", 2))
````

## Challenge Exercise: Mini Data Pipeline

Combine everything from this module. Build a mini data pipeline:

1. Given `raw_records` (list of dicts with 'age', 'income', 'label')
2. Write `clean_record(record)` that:
   - Converts 'age' and 'income' to float (use `safe_convert`)
   - Normalizes 'label' string (strip, lowercase)
   - Returns `None` if age or income is missing after conversion
3. Write `process_dataset(records)` that applies cleaning and returns only valid records
4. Print summary: total records, valid, invalid, and average age of valid records

````python
raw_records = [
    {"age": "25", "income": "50000", "label": " Positive "},
    {"age": "N/A", "income": "60000", "label": "negative"},
    {"age": "30", "income": "?", "label": "Positive"},
    {"age": "45", "income": "80000", "label": " NEGATIVE "},
    {"age": "22", "income": "35000", "label": "positive"},
]

# YOUR CODE HERE
````

````python
# SOLUTION
def safe_convert(value, dtype=float, default=None):
    try:
        return dtype(value)
    except (ValueError, TypeError):
        return default

def clean_record(record):
    age = safe_convert(record.get("age"))
    income = safe_convert(record.get("income"))
    if age is None or income is None:
        return None
    return {"age": age, "income": income, "label": record["label"].strip().lower()}

def process_dataset(records):
    cleaned = [clean_record(r) for r in records]
    valid = [r for r in cleaned if r is not None]
    invalid = len(records) - len(valid)
    avg_age = sum(r["age"] for r in valid) / len(valid) if valid else 0
    print(f"Total: {len(records)} | Valid: {len(valid)} | Invalid: {invalid}")
    print(f"Average age (valid): {avg_age:.1f}")
    return valid

raw_records = [
    {"age": "25", "income": "50000", "label": " Positive "},
    {"age": "N/A", "income": "60000", "label": "negative"},
    {"age": "30", "income": "?", "label": "Positive"},
    {"age": "45", "income": "80000", "label": " NEGATIVE "},
    {"age": "22", "income": "35000", "label": "positive"},
]

valid_records = process_dataset(raw_records)
for r in valid_records:
    print(f"  {r}")
````
