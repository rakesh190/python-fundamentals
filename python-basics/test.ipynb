```python
movies = [
    {"title": "Inception",       "rating": 8.8, "genre": "sci-fi",  "year": 2010},
    {"title": "The Dark Knight", "rating": 9.0, "genre": "action",  "year": 2008},
    {"title": "Interstellar",    "rating": 8.6, "genre": "sci-fi",  "year": 2014},
    {"title": "Parasite",        "rating": 8.5, "genre": "drama",   "year": 2019},
    {"title": "Dune",            "rating": 7.9, "genre": "sci-fi",  "year": 2021},
    {"title": "1917",            "rating": 8.2, "genre": "drama",   "year": 2019},
    {"title": "Joker",           "rating": 8.4, "genre": "drama",   "year": 2019},
    {"title": "Tenet",           "rating": 7.3, "genre": "action",  "year": 2020},
]
```

```python
#Extract all movie titles
titles = [m['title'] for m in movies]
print(titles)
```

```python
#Filter highly rated movies
high_rated_movies = [m['title'] for m in movies if m['rating']> 8.5]
print(high_rated_movies)
```

```python
# print a numbered watchlist
for i, m in enumerate(movies, start=1):
  print(f"#{i}  {m['title']}  {m['rating']}")
```

```python
#Build a dict where each key is a genre and the value is the average rating for that genre. Then print each genre and its average.
genre_dict = {}

for genre, rating in [(m['genre'], m['rating']) for m in movies]:
    if genre in genre_dict:
        genre_dict[genre].append(rating)
    else:
        genre_dict[genre] = [rating]

for genre, ratings in genre_dict.items():
    avg = round(sum(ratings) / len(ratings), 2)
    print(f"{genre:<10} -> avg rating: {avg}")
  



```

```python
#Write a function convert_temp(value, unit="C") that converts temperature. If unit is "C" convert Celsius → Fahrenheit. If unit is "F" convert Fahrenheit → Celsius. Round to 1 decimal.

def convert_temp(value, unit='C'):
  if(unit.lower()== 'f'):
    temp =  (value * 9/5) + 32
    return f"{round(temp, 1)}°F"
  if(unit.lower() == 'c'):
    temp =  (value -32) * 5/9
    return f"{round(temp, 1)}°C"
  
print(convert_temp(35, 'f'))
```

```python
# Given the list below, produce three sorted versions using sorted() and a lambda for each.


employees = [
    {"name": "Alice",   "age": 32, "salary": 85000},
    {"name": "Bob",     "age": 25, "salary": 62000},
    {"name": "Charlie", "age": 41, "salary": 110000},
    {"name": "Diana",   "age": 28, "salary": 74000},
]
 
 #sort by age
sorted_by_age = sorted(employees, key=lambda e:e['age']);
print(sorted_by_age)

#sort by salary
sorted_by_salary = sorted(employees, key=lambda e: e['salary']);
print(sorted_by_salary)

#sort by name length
sorted_by_name_len = sorted(employees, key=lambda e: len(e['name']))
print(sorted_by_name_len)
```

```python
#Write a function stats(*numbers) that accepts any number of values and returns the count, sum, min, max, and average — all in one return statement. Then unpack and print them.

def stats(*numbers):
  num_count = len(numbers)
  number_sum = sum(numbers)
  min_num = min(numbers)
  max_num = max(numbers)
  avg = round(number_sum/len(numbers))
  return (num_count, number_sum, min_num, max_num, avg)

num_count, number_sum, min_sum, max_num, avg = stats(1,1,2,2,3,3)
print(num_count)
print(number_sum);
print(min_sum)
print(max_num)
print(avg)

```

```python
#Write a function apply_discount(price, **options) that applies a discount to a price. The options dict can have: percent (discount %), label (name of deal), and min_price (floor — price never goes below this). All are optional.

def apply_discount(price, **options):
  percent = options.get('percent', 0)
  label= options.get('label', 'Original')
  min_price = options.get('min_price', 0)
  print(percent, label, min_price)

  final = price * (1- percent / 100)
  final = max(final, min_price)
  print(f"{label}: ${price} → Final: ${final}") 

apply_discount(25, label='flash', percent=30, min_price=35)
  
```
