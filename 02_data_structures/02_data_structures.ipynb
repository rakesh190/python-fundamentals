## 1. Lists — Ordered, Mutable Sequences

````python
# Lists in ML: storing epochs, losses, predictions, feature names
feature_names = ['age', 'income', 'education_years', 'credit_score']
train_losses  = [1.2, 0.98, 0.76, 0.61, 0.52, 0.45, 0.40]

# Indexing & slicing
print(f"First feature: {feature_names[0]}")
print(f"Last feature:  {feature_names[-1]}")
print(f"First 5 losses: {train_losses[:5]}")
print(f"Losses reversed: {train_losses[::-1]}")

# Common list operations
feature_names.append('spending_score')   # add one
feature_names.extend(['gender', 'city']) # add many
feature_names.remove('city')             # remove by value
popped = feature_names.pop()             # remove last
print(f"Features after edits: {feature_names}")
print(f"Popped: {popped}")
````

````python
# Sorting — important for ranking features by importance
importances = [0.12, 0.34, 0.08, 0.25, 0.21]
features    = ['age', 'income', 'city', 'credit', 'education']

# Sort by importance descending
ranked = sorted(zip(features, importances), key=lambda x: x[1], reverse=True)
print("Feature Importance Ranking:")
for rank, (feat, imp) in enumerate(ranked, 1):
    print(f"  {rank}. {feat:<12} {imp:.2f}")
````

## 3. Dictionaries — Key-Value Storage

````python
# Dicts: hyperparameters, metrics, configs — used everywhere in ML
hyperparams = {
    'n_estimators': 100,
    'max_depth': 5,
    'learning_rate': 0.1,
    'subsample': 0.8
}

# Safe access
depth = hyperparams.get('max_depth', 3)          # default=3
reg = hyperparams.get('regularization', None)    # missing key → None
print(f"max_depth={depth}, regularization={reg}")

# Update / merge configs
overrides = {'max_depth': 8, 'n_estimators': 200}
final_params = {**hyperparams, **overrides}      # merge, override wins
print(f"\nFinal params: {final_params}")

# Iteration patterns
print("\nHyperparameter summary:")
for key, value in final_params.items():
    print(f"  {key:<20}: {value}")
````

````python
# defaultdict — accumulating results per class
from collections import defaultdict

predictions = [
    ('cat', 'cat'), ('dog', 'cat'), ('bird', 'bird'),
    ('cat', 'cat'), ('dog', 'dog'), ('bird', 'cat')
]

class_correct = defaultdict(int)
class_total   = defaultdict(int)

for true_label, pred_label in predictions:
    class_total[true_label] += 1
    if true_label == pred_label:
        class_correct[true_label] += 1

print("Per-class accuracy:")
for cls in class_total:
    acc = class_correct[cls] / class_total[cls]
    print(f"  {cls}: {acc:.0%} ({class_correct[cls]}/{class_total[cls]})")
````

## 5. Comprehensions — Pythonic Data Processing

````python
import math

# List comprehension — apply transformations
raw_values = [-2, 0, 3, -1, 5, 2, -4, 1]
relu_values = [max(0, x) for x in raw_values]              # ReLU activation
log_values  = [math.log(x) for x in raw_values if x > 0]  # log of positives only
squared     = [x**2 for x in raw_values]                   # squared

print(f"ReLU:    {relu_values}")
print(f"Log(+):  {[round(v,3) for v in log_values]}")
print(f"Squared: {squared}")
````

````python
# Dict comprehension — build lookup tables
classes = ['cat', 'dog', 'bird', 'fish']
class_to_idx = {cls: idx for idx, cls in enumerate(classes)}
idx_to_class = {idx: cls for cls, idx in class_to_idx.items()}

print(f"class→idx: {class_to_idx}")
print(f"idx→class: {idx_to_class}")

# Encode predictions
pred_classes = ['dog', 'cat', 'bird', 'dog']
pred_encoded = [class_to_idx[c] for c in pred_classes]
print(f"Encoded:   {pred_encoded}")
````

````python
# Nested comprehension — confusion matrix skeleton
n_classes = 3
confusion = [[0] * n_classes for _ in range(n_classes)]  # 3x3 matrix of zeros

test_data = [(0,0), (1,2), (2,2), (0,1), (1,1), (2,0)]  # (true, pred)
for true, pred in test_data:
    confusion[true][pred] += 1

print("Confusion Matrix:")
for row in confusion:
    print(f"  {row}")
````

## 6. Useful Built-ins: zip, enumerate, map, filter

````python
features = ['age', 'income', 'score']
values   = [25, 52000, 0.87]
weights  = [0.3, 0.5, 0.2]

# zip — pair features with values
feature_dict = dict(zip(features, values))
print(f"Feature dict: {feature_dict}")

# Weighted sum using zip
weighted_sum = sum(v * w for v, w in zip(values, weights))
print(f"Weighted sum: {weighted_sum:.2f}")

# enumerate — epoch tracking
losses = [0.9, 0.7, 0.55, 0.42, 0.35]
for epoch, loss in enumerate(losses, start=1):
    print(f"  Epoch {epoch}: {loss:.2f}")

# map & filter
raw = ['1.2', '3.4', '?', '5.6', 'N/A']
numeric = list(filter(None, map(lambda x: float(x) if x.replace('.','').isdigit() else None, raw)))
print(f"\nFiltered numeric: {numeric}")
````

## 7. Generators — Memory-Efficient Iteration

````python
# Generator function — simulate data batching
def batch_generator(data, batch_size=3):
    """Yield batches from data (like PyTorch DataLoader)."""
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]

dataset = list(range(10))  # 10 samples
print("Batches:")
for batch_idx, batch in enumerate(batch_generator(dataset, batch_size=3)):
    print(f"  Batch {batch_idx}: {batch}")

# Generator expression — memory efficient vs list
n = 1_000_000
gen_sum = sum(x**2 for x in range(n))  # no list created in memory
print(f"\nSum of squares: {gen_sum:,}")
````

## Challenge: Build a Mini Vocabulary Builder

NLP tasks need vocabularies. Implement:

1. `build_vocab(corpus, max_size=None)` — takes a list of sentences, returns `{word: index}` dict (sorted by frequency, most common first, starting index at 1; index 0 reserved for `<UNK>`)
2. `encode(sentence, vocab)` — converts sentence to list of indices (unknown words → 0)
3. `decode(indices, vocab)` — converts indices back to words

Test corpus:
```python
corpus = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "a cat and a dog"
]
```

````python
corpus = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "a cat and a dog"
]
# YOUR CODE HERE
````

````python
# SOLUTION
from collections import Counter

def build_vocab(corpus, max_size=None):
    words = [word for sentence in corpus for word in sentence.split()]
    counts = Counter(words)
    sorted_words = [w for w, _ in counts.most_common(max_size)]
    vocab = {'<UNK>': 0}
    vocab.update({word: idx+1 for idx, word in enumerate(sorted_words)})
    return vocab

def encode(sentence, vocab):
    return [vocab.get(word, 0) for word in sentence.split()]

def decode(indices, vocab):
    idx_to_word = {v: k for k, v in vocab.items()}
    return [idx_to_word.get(i, '<UNK>') for i in indices]

corpus = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "a cat and a dog"
]

vocab = build_vocab(corpus)
print(f"Vocab ({len(vocab)} words): {vocab}")

test_sentence = "the cat chased the dog"
encoded = encode(test_sentence, vocab)
decoded = decode(encoded, vocab)
print(f"\nOriginal: {test_sentence}")
print(f"Encoded:  {encoded}")
print(f"Decoded:  {decoded}")
````
