## 2. Indexing, Slicing & Boolean Masking

````python
np.random.seed(42)
X = np.random.randn(6, 4)  # 6 samples, 4 features
print("Dataset X:")
print(X.round(2))

# Basic indexing
print(f"\nRow 0: {X[0].round(2)}")
print(f"Col 2: {X[:, 2].round(2)}")
print(f"Rows 1-3: \n{X[1:4].round(2)}")
print(f"Element [2,1]: {X[2,1]:.4f}")
````

````python
# Boolean masking — critical for data filtering
ages = np.array([22, 45, 17, 31, 28, 15, 52, 29])
incomes = np.array([35000, 80000, 0, 60000, 45000, 0, 95000, 55000])

# Filter adults with income
adult_mask = ages >= 18
employed_mask = incomes > 0
valid_mask = adult_mask & employed_mask

print(f"Adult mask:    {adult_mask}")
print(f"Employed mask: {employed_mask}")
print(f"Valid ages:    {ages[valid_mask]}")
print(f"Valid incomes: {incomes[valid_mask]}")

# Fancy indexing
selected = np.array([0, 2, 5])   # pick specific samples
print(f"\nSelected ages: {ages[selected]}")
````

````python
# np.where — vectorized if-else (used for clipping, encoding)
scores = np.array([0.3, 0.7, 0.6, 0.4, 0.9, 0.1, 0.55])
threshold = 0.5

predictions = np.where(scores >= threshold, 1, 0)
print(f"Scores:      {scores}")
print(f"Predictions: {predictions}")

# Clip outliers
raw = np.array([-5, 2, 100, 3, -200, 4, 50])
clipped = np.clip(raw, -10, 10)
print(f"\nClipped: {clipped}")
````

````python
# Element-wise operations
a = np.array([1, 2, 3, 4, 5], dtype=float)
b = np.array([2, 2, 2, 2, 2], dtype=float)

print(f"a + b = {a + b}")
print(f"a * b = {a * b}")
print(f"a ** 2 = {a ** 2}")
print(f"np.sqrt(a) = {np.sqrt(a).round(3)}")
print(f"np.exp(a) = {np.exp(a).round(2)}")
print(f"np.log(a) = {np.log(a).round(3)}")
````

````python
# Broadcasting — key concept for ML operations
# Subtract mean from each feature (centering)
np.random.seed(42)
X = np.random.randn(5, 3)   # 5 samples, 3 features
means = X.mean(axis=0)      # shape (3,) — mean per feature
stds  = X.std(axis=0)       # shape (3,) — std per feature

X_scaled = (X - means) / stds  # broadcasting: (5,3) - (3,) → (5,3)

print("Original X (first 3 rows):")
print(X[:3].round(3))
print("\nZ-scored X (first 3 rows):")
print(X_scaled[:3].round(3))
print(f"\nMean after scaling (≈0): {X_scaled.mean(axis=0).round(10)}")
print(f"Std after scaling (≈1):  {X_scaled.std(axis=0).round(10)}")
````

## 4. Linear Algebra — The Heart of ML

````python
# Dot product & matrix multiplication
# Linear layer: y = X @ W + b
np.random.seed(0)
X = np.random.randn(4, 3)   # 4 samples, 3 input features
W = np.random.randn(3, 2)   # weight matrix: 3 inputs → 2 outputs
b = np.zeros(2)             # bias

output = X @ W + b          # matrix multiply + broadcast add
print(f"X shape: {X.shape}")
print(f"W shape: {W.shape}")
print(f"Output shape: {output.shape}")
print(f"Output:\n{output.round(3)}")
````

````python
# Key linear algebra operations
A = np.array([[2., 1.], [5., 3.]])

det = np.linalg.det(A)              # determinant
inv = np.linalg.inv(A)              # inverse
eigenvalues, eigenvectors = np.linalg.eig(A)  # eigendecomposition (PCA)

print(f"Determinant: {det}")
print(f"Inverse:\n{inv}")
print(f"A @ A_inv (should be I):\n{(A @ inv).round(10)}")
print(f"Eigenvalues: {eigenvalues.round(4)}")
````

````python
# Norms — used in regularization, gradient clipping
v = np.array([3., 4., 0.])

l1_norm = np.linalg.norm(v, ord=1)   # L1 (Manhattan)
l2_norm = np.linalg.norm(v, ord=2)   # L2 (Euclidean)
linf_norm = np.linalg.norm(v, ord=np.inf)  # L∞

print(f"Vector: {v}")
print(f"L1 norm: {l1_norm}")
print(f"L2 norm: {l2_norm}")
print(f"L∞ norm: {linf_norm}")
````

Steps:

4. Compare `theta_hat` to `true_theta`

````python
# SOLUTION
np.random.seed(42)
X_raw = np.random.randn(50, 2)
true_theta = np.array([3.0, -1.5, 2.0])

# 1. Add bias column
X = np.column_stack([np.ones(50), X_raw])  # shape (50, 3)

# 2. Generate y with noise
noise = np.random.randn(50) * 0.5
y = X @ true_theta + noise

# 3. Normal equation
theta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

# 4. Compare
print(f"True theta:     {true_theta}")
print(f"Estimated theta:{theta_hat.round(4)}")
print(f"Difference:     {(theta_hat - true_theta).round(4)}")

# Predictions and MSE
y_pred = X @ theta_hat
mse = np.mean((y - y_pred)**2)
print(f"MSE: {mse:.4f}")
````

### Exercise 3.4 — Softmax & Cross-Entropy
Implement these two ML functions using only NumPy:

**Softmax:** `softmax(z)_i = exp(z_i) / sum(exp(z_j))`  (numerically stable: subtract max first)

**Cross-Entropy Loss:** `L = -sum(y_true * log(y_pred + eps))`

Test with:
```python
logits = np.array([[2.0, 1.0, 0.1], [0.5, 2.5, 0.3], [0.1, 0.2, 3.0]])
y_true = np.array([[1,0,0], [0,1,0], [0,0,1]])  # one-hot
```

````python
logits = np.array([[2.0, 1.0, 0.1], [0.5, 2.5, 0.3], [0.1, 0.2, 3.0]])
y_true = np.array([[1,0,0], [0,1,0], [0,0,1]])
# YOUR CODE HERE
````

````python
# SOLUTION
def softmax(z):
    """Numerically stable softmax."""
    z_stable = z - z.max(axis=1, keepdims=True)  # subtract max per row
    exp_z = np.exp(z_stable)
    return exp_z / exp_z.sum(axis=1, keepdims=True)

def cross_entropy_loss(y_true, y_pred, eps=1e-9):
    """Mean cross-entropy loss over batch."""
    return -np.mean(np.sum(y_true * np.log(y_pred + eps), axis=1))

logits = np.array([[2.0, 1.0, 0.1], [0.5, 2.5, 0.3], [0.1, 0.2, 3.0]])
y_true = np.array([[1,0,0], [0,1,0], [0,0,1]])

probs = softmax(logits)
loss = cross_entropy_loss(y_true, probs)

print("Logits:")
print(logits)
print("\nSoftmax probabilities:")
print(probs.round(4))
print(f"\nRow sums (should be 1): {probs.sum(axis=1)}")
print(f"Cross-entropy loss: {loss:.4f}")
````

## Challenge: Implement Gradient Descent from Scratch

Implement gradient descent for **logistic regression** on a 2D dataset:

1. Generate `X` (100×2), `y` (binary 0/1) with `make_blobs` style using numpy
2. Add bias column
3. Implement `sigmoid(z)`, `predict_proba(X, theta)`, `compute_loss(y, y_hat)` (binary cross-entropy)
4. Implement gradient: `grad = X.T @ (y_hat - y) / n`
5. Train for 200 epochs with lr=0.1, print loss every 50 epochs
6. Report final accuracy

````python
np.random.seed(42)
# Generate separable 2-class dataset
n = 100
X_class0 = np.random.randn(n//2, 2) + np.array([-2, -2])
X_class1 = np.random.randn(n//2, 2) + np.array([2, 2])
X_raw = np.vstack([X_class0, X_class1])
y = np.hstack([np.zeros(n//2), np.ones(n//2)])

# YOUR CODE HERE
````

````python
# SOLUTION
np.random.seed(42)
n = 100
X_class0 = np.random.randn(n//2, 2) + np.array([-2, -2])
X_class1 = np.random.randn(n//2, 2) + np.array([2, 2])
X_raw = np.vstack([X_class0, X_class1])
y = np.hstack([np.zeros(n//2), np.ones(n//2)])

# Add bias
X = np.column_stack([np.ones(n), X_raw])
theta = np.zeros(3)

def sigmoid(z):
    return 1 / (1 + np.exp(-np.clip(z, -500, 500)))

def predict_proba(X, theta):
    return sigmoid(X @ theta)

def compute_loss(y, y_hat, eps=1e-9):
    return -np.mean(y * np.log(y_hat + eps) + (1 - y) * np.log(1 - y_hat + eps))

# Training loop
lr = 0.1
for epoch in range(200):
    y_hat = predict_proba(X, theta)
    loss = compute_loss(y, y_hat)
    grad = X.T @ (y_hat - y) / n
    theta -= lr * grad
    if epoch % 50 == 0:
        print(f"Epoch {epoch:3d}: loss={loss:.4f}")

# Final accuracy
final_preds = (predict_proba(X, theta) >= 0.5).astype(int)
accuracy = np.mean(final_preds == y)
print(f"\nFinal accuracy: {accuracy:.2%}")
print(f"Learned theta: {theta.round(4)}")
````
